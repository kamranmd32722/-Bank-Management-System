```python
a = 15
b = 4

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Floor Division:", a // b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

Step 1️⃣: Values assign karna
a = 15
b = 4

Yaha do variables banaye gaye:

a me 15 store kiya

b me 4 store kiya

Step 2️⃣: Arithmetic operations perform karna

Program a aur b par different maths operations laga raha hai.

🔢 Har operator ka logic:

1️⃣ Addition (+)
15 + 4 = 19
👉 Dono numbers ko add karta hai.

2️⃣ Subtraction (-)
15 - 4 = 11
👉 Pehle number me se dusra minus karta hai.

3️⃣ Multiplication (*)
15 × 4 = 60
👉 Dono numbers ko multiply karta hai.

4️⃣ Division (/)
15 ÷ 4 = 3.75
👉 Exact division karta hai, decimal answer deta hai.

5️⃣ Floor Division (//)
15 ÷ 4 = 3
👉 Decimal hata deta hai (sirf integer part leta hai).

6️⃣ Modulus (%)
15 % 4 = 3
👉 Division ke baad jo remainder bachta hai, wo deta hai.

7️⃣ Power (**)
15⁴ = 50625
👉 15 ko 4 times multiply karta hai (15×15×15×15).

```python
num = 20
print("Initial:", num)

num += 5
print("After += :", num)

num -= 3
print("After -= :", num)

num *= 2
print("After *= :", num)

num //= 4
print("After //= :", num)

```

Step 1️⃣: Initial value
num = 20

👉 num me starting value 20 store hai.

Output:

Initial: 20
Step 2️⃣: Assignment Operators ka use

Yeh program compound assignment operators use karta hai.
Ye operators shortcut hote hain:

num += 5 ka matlab → num = num + 5

🔎 Step-by-step Calculation

1️⃣ num += 5
20 + 5 = 25

After += : 25

2️⃣ num -= 3
25 - 3 = 22

After -= : 22

3️⃣ num *= 2
22 × 2 = 44

After *= : 44

4️⃣ num //= 4
44 ÷ 4 = 11 (floor division)

After //= : 11# New Section
